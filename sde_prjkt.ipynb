```python
import numpy as np
import matplotlib.pyplot as plt
import cmath
```

```python
# lambda1 = 3, lambda2=30
```

```python
def potential(x1, x2):
  return 1.5*x1**2 + 3*x2**2
```

```python
np.linalg.eig(np.diag((2,4)))
```

```python
def w0(t):
  return 2*np.exp(-0.75*t), 4*np.exp(-0.75*10*t)
```

```python
mu = 4
m = 2
```

```python
def get_eigen(k):
  return -mu/(2*m) + cmath.sqrt((mu/m)*(mu/(4*m) - k/mu)), -mu/(2*m) - cmath.sqrt((mu/m)*(mu/(4*m) - k/mu))
```

```python
lam1 = get_eigen(3)
```

```python
lam2 = get_eigen(30)
```

```python
def w(t):
  return (np.exp(lam1[0]*t) +  np.exp(lam1[1]*t)).real, (2*np.exp(lam2[0]*t) +  2*np.exp(lam2[1]*t)).real
```

```python
x = np.linspace(0,5,1000)
```

```python
w_mom = np.array([w(i) for i in x])
w_no_mom = np.array([w0(i) for i in x])
```

```python
plt.plot(x, w_mom[:,0], 'r')
plt.plot(x, w_no_mom[:,0], 'g')
plt.grid()
plt.ylabel("W1")
plt.xlabel("Time")
plt.title("Case where condition is met")
```

```python
plt.plot(x, w_mom[:,1], 'r')
plt.plot(x, w_no_mom[:,1], 'g')
plt.grid()
plt.ylabel("W2")
plt.xlabel("Time")
plt.title("Case where condition is not met")
```

```python
def w_discrete(w1, w2, momentum, epsilon):
  return (1+momentum -epsilon*3)*w2[0] - momentum*w1[0], (1+momentum -epsilon*30)*w2[1] - momentum*w1[1]
```

```python
delta_t = 0.5
ep,p = (delta_t**2)/(m+mu*delta_t),m/(m+mu*delta_t)
print(ep,p)
disc_w = [(2,4), (2,4)]
for i in range(int(5/delta_t)):
  disc_w.append(
      w_discrete(disc_w[-2], disc_w[-1], p, ep)
  )
disc_w = np.array(disc_w)
plt.plot([delta_t*i for i in range(int(5/delta_t)+1)], disc_w[1:,0], label="weight 1", color='b', marker="*")
plt.plot([delta_t*i for i in range(int(5/delta_t)+1)], disc_w[1:,1], label="weight 2", color='g', marker="*")
plt.grid()
plt.ylabel("Weights")
plt.xlabel("Time")
plt.title(f"Delta_t = {delta_t}")
plt.legend()
```
